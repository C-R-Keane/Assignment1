```python
# Import packages
from bs4 import BeautifulSoup
import urllib.request
import csv
import pandas as pd
```

```python
#create list to hold reviews 
written_review =[]
star_review =[]
userID = []

#loop through 10 pages (for 100 reviews or less)
for i in range(10):
    # assign website URL
    url = 'https://www.yelp.com/biz/emack-and-bolios-new-york-4?osq=Emack+and+Bolio%27s&start='+str(10*i)

    # OpenURL
    ourUrl = urllib.request.urlopen(url)

    # BeautifulSoup object which represents the document as a nested data structure
    soup = BeautifulSoup(ourUrl,'html.parser')    
    
    # find the written reviews for one page 
    for i in soup.find_all('p',{'class':'comment__373c0__Nsutg css-n6i4z7'}):
        individual_review = i.text 
        written_review.append(individual_review)
    
    #For the userID
    for i in soup.find_all('span', {'class':'fs-block css-m6anxm'}):
        individual_userid = i.text
        
        if individual_userid != 'Username': 
            userID.append(individual_userid)
            
    #for loop that will assign the stars given         
    for i in soup.find_all('div',{'class':'review__373c0__3MsBX border-color--default__373c0__1WKlL'}):
        #identifies the span that is needed
        identify_review = i

        #loop to append the aria-label
        for k in identify_review.find('span',{'class':'display--inline__373c0__1gaV4 border-color--default__373c0__1yxBb'}):
            #identifies the span that is needed
            identify_star = (k['aria-label'])
            star_review.append(identify_star)
 
```

```python
print(len(written_review))# identifies the total reviews
print(len(userID))# identifies the total users
print(len(star_review))# identifies the total number stars
```

```python
# create a dataframe based on the lists created above
ice_cream = pd.DataFrame(list(zip(userID, star_review, written_review)),
                  columns = ['ReviewID', 'ReviewRating', 'ReviewText'])

ice_cream.to_csv('ice_cream_shop_reviews.csv', encoding='utf-8')
```

```python
# The folllowing is the Term-Document Matrix for 6 Key Words 

matrix_key = ["awesome", "overpriced", "bad", "flavors", "Love", "Decent"]

awesome = "awesome"
friendly = "friendly "
love = "love"
bad = "bad"
overpriced = "overpriced"
negative = "negative"

#append all information to this list for future columns 
matrix_key_list =[]
matrix_count = []

# count how often any of the key words are in each review 
for i in written_review:
    #Check to see if the Matrix-Key words are present, if not do nothing 
    row_text = i.lower()
    
    # break it out by word, to ensure that columns will look nice 
    if awesome in row_text:
        awesome_count = row_text.count(awesome)
    else:
        awesome_count = "0"
        
    if friendly in row_text:
        friendly_count = row_text.count(friendly) 
    else:
        friendly_count = "0"
        
    if love in row_text:
        love_count = row_text.count(love)
    else:
        love_count = "0"
        
    if bad in row_text:
        bad_count = row_text.count(bad)
    else:
        bad_count = "0"
        
    if overpriced in row_text:
        overpriced_count = row_text.count(overpriced) 
    else:
        overpriced_count = "0"
        
    if negative in row_text:
        negative_count = row_text.count(negative)
    else:
        negative_count = "0"    
    
    
    # make each count into a row
    row = [row_text, awesome_count, friendly_count, love_count, bad_count, overpriced_count, negative_count]
    
    matrix_key_list.append(row)

# create the dataframe
Matrix_key = pd.DataFrame(list(matrix_key_list),
                  columns = ['ReviewText', 'Awesome', 'Friendly', 
                             'Love', 'Bad', 'Overpriced', 'Negative'])
# send to csv
Matrix_key.to_csv('Term_Doc_Matrix.csv', encoding='utf-8')
```
